```python
 # aggregated transaction

path1 = r'D:\phone pe\pulse\data\aggregated\transaction\country\india\state/'
agg_state_list = os.listdir(path1)

clm1 = {
    'State': [],'Year': [],'Quarter': [],'Transaction_type': [],'Transaction_count': [],'Transaction_amount': []
}

for state in agg_state_list:
    clm_state= path1 + state +"/"
    agg_yr = os.listdir(clm_state)

    for year in agg_yr:
        clm_year = clm_state + year + "/"
        agg_year_list = os.listdir(clm_year)

        for file in agg_year_list:
            clm_file = clm_year + file
            data = open(clm_file, 'r')
            d = json.load(data)

    
            for i in d["data"]["transactionData"]:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                clm1['Transaction_type'].append(name)
                clm1['Transaction_count'].append(count)
                clm1['Transaction_amount'].append(amount)
                clm1['State'].append(state)
                clm1['Year'].append(year)
                clm1['Quarter'].append(int(file.strip('.json')))


# Successfully created a DataFrame
agg_trans1 = pd.DataFrame(clm1)


agg_trans1['State']=agg_trans1['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
agg_trans1['State']=agg_trans1['State'].str.replace('dadra-&-nagar-haveli-&-daman-&-diu','dadra and nagar haveli and daman and diu')
agg_trans1['State']=agg_trans1['State'].str.replace("-"," ")
agg_trans1['State'] = agg_trans1['State'].str.title()
```
